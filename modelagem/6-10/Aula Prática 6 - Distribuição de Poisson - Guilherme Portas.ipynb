A distribuição de poissom é ótima caso tenha eventos raros e improváveis em intervalos contínuos, enquanto a binomial é boa para contagem de sucessos com um número fixo de tentativas. Além disso, a Poisson pode ser uma aproximação da binomial quando o número de tentativas é grande para uma probabilidade pequena.

```python
# 1)
def plot_poisson_distribution(lambda_value, n_samples=1000):
    dados_poisson = numpy.random.poisson(lambda_value, n_samples)
    mu = numpy.mean(dados_poisson)
    sigma = numpy.std(dados_poisson)
    plt.figure(figsize=(12, 6))
    sns.histplot(dados_poisson, bins=30, kde=False, color='blue', stat='density', label='Dados Poisson', alpha=0.6)
    
    x = numpy.linspace(min(dados_poisson), max(dados_poisson), 100)
    y = norm.pdf(x, mu, sigma)
    plt.plot(x, y, color='red', label='Curva Normal Ajustada', linewidth=2)
    plt.title(f'Histograma da Distribuição de Poisson (λ = {lambda_value})')
    plt.xlabel('Valores')
    plt.ylabel('Densidade')
    plt.legend()
    plt.show()

lambda_values = [2, 5, 10]

for lambda_value in lambda_values:
    plot_poisson_distribution(lambda_value)
```

```python
# 2)
# normal
n = 1000
mu = 0
sigma = 1
dados_normais = numpy.random.normal(mu, sigma, n)

# binomial
n_bin = 10
p_bin = 0.5
dados_binomiais = numpy.random.binomial(n_bin, p_bin, n)

# Poisson
lambda_poisson = 5
dados_poisson = numpy.random.poisson(lambda_poisson, n)

plt.figure(figsize=(12, 6))
sns.histplot(dados_normais, kde=True, color='blue', label='Normal', stat='density', bins=30)
sns.histplot(dados_binomiais, kde=True, color='orange', label='Binomial', stat='density', bins=30)
sns.histplot(dados_poisson, kde=True, color='green', label='Poisson', stat='density', bins=30)
plt.legend()
plt.title('Comparação das Distribuições')
plt.xlabel('Valor')
plt.ylabel('Densidade')
plt.show()


```
